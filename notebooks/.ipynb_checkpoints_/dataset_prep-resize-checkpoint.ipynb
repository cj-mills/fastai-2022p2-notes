**Define funtion that resizes an image and saves the result to disk**

```python
def array_func(img_id, img_path_dict, save_dir, target_sz=512, divisor=32, img_fmt='JPEG', quality=100):
    """
    Resizes an image to a specified size and saves it to a specified directory.

    Parameters:
    img_id (int): The ID of the image to resize.
    img_path_dict (dict): A dictionary that maps image IDs to file paths.
    save_dir (Path): The directory to save the resized image to.
    target_sz (int, optional): The target size of the resized image. Defaults to 512.
    divisor (int, optional): The divisor to use for resizing. Defaults to 32.
    img_fmt (str, optional): The format to save the image in. Defaults to 'JPEG'.
    quality (int, optional): The quality of the saved image. Defaults to 100.

    Returns:
    None
    """
    img_path = img_path_dict[img_id]
    
    # Check if the image file exists
    if not img_path_dict[img_id].exists():
        return
    
    # Open the image
    img = Image.open(img_path)
    
    # Resize the image
    resized_img = resize_img(img, target_sz=target_sz, divisor=divisor)
    
    # Determine the save path for the resized image
    save_path = Path(save_dir/f"{img_id}.{img_fmt.lower()}")

    # Save the resized image
    resized_img.save(save_path, img_fmt, quality=quality)
```
