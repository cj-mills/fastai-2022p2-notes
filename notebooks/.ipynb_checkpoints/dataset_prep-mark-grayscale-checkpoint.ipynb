**Map image IDs to file paths**

```python
# Create a dictionary mapping image IDs to image paths
img_path_dict = {int(path.stem.split('-')[-1]) : path for path in img_paths}

# Display the first few entries in the dictionary using a Pandas Dataframe
pd.DataFrame.from_dict(img_path_dict, orient='index').head()
```

**Get image IDs**

```python
img_ids = list(img_path_dict.keys())
len(img_ids)
```

```python
def is_mostly_grayscale(img, threshold=0.9, std_threshold=3):
    
    img_array = np.array(img)

    # Compute the standard deviation of the RGB values across each pixel
    std_dev = np.std(img_array, axis=2)

    # Compute the number of grayscale pixels (i.e., pixels where the standard deviation is low)
    num_grayscale = np.sum(std_dev < std_threshold)

    # Compute the total number of pixels in the img_array
    total_pixels = img_array.shape[0] * img_array.shape[1]

    # Compute the percentage of grayscale pixels
    percent_grayscale = num_grayscale / total_pixels

    # Check if the percentage of grayscale pixels is above the threshold
    return percent_grayscale > threshold
```

**Define funtion that resizes an image and saves the result to disk**

```python
def array_func(img_id, img_path_dict, img_id_queue):
    """
    Resizes an image to a specified size and saves it to a specified directory.

    Parameters:
    img_id (int): The ID of the image to resize.
    img_path_dict (dict): A dictionary that maps image IDs to file paths.
    save_dir (Path): The directory to save the resized image to.
    target_sz (int, optional): The target size of the resized image. Defaults to 512.
    divisor (int, optional): The divisor to use for resizing. Defaults to 32.
    img_fmt (str, optional): The format to save the image in. Defaults to 'JPEG'.
    quality (int, optional): The quality of the saved image. Defaults to 100.

    Returns:
    None
    """
    img_path = img_path_dict[img_id]
    
    # Check if the image file exists
    if not img_path_dict[img_id].exists():
        return
    
    # Open the image
    img = Image.open(img_path)
    
    if is_mostly_grayscale(img, threshold=0.9, std_threshold=6):
        img_id_queue.put(img_id)
```
