**Open sample image**

```python
image = Image.open('pexels-photo-567540.jpeg').convert('RGB')
image
```

**Convert image to NumPy array**

```python
image_array = np.array(image)
image_array.shape
```

**Generate noise**

```python
noise = np.random.normal(0, 1, image_array.shape)*80
Image.fromarray(noise.astype(np.uint8))
```

**Add noise to image**

```python
noisy_image = image_array + noise
Image.fromarray(noisy_image.astype(np.uint8))
```

**Subtract noise from image**

```python
denoised_image = noisy_image - noise
Image.fromarray(denoised_image.astype(np.uint8))
```
